Q1. Handle Missing Values (Basic Cleaning)
Fill missing values using the following:
Age → Median
Salary → Mean
Hours_Worked_Per_Week → Median
Performance_Score → Mean
Display the dataset after handling missing values

```python
import pandas as pd
df=pd.read_csv('employee_productivity_dataset.csv')
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median(),inplace=True)
df['Performance_Score'].fillna(df['Performance_Score'].mean(),inplace=True)
print(df.head())
```

Q2. Label Encoding
Convert these columns into numeric values using Label Encoding:
Gender
Department
Show the updated columns.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Department']=le.fit_transform(df['Department'])
df[['Gender','Department']]
```

Q3. One-Hot Encoding
Apply One-Hot Encoding on:
Work_Mode
Location
Display the dataset and check how many new columns are created.

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
ohe = OneHotEncoder(sparse_output=False)
encoded_data = ohe.fit_transform(df[['Work_Mode', 'Location']])
new_columns = ohe.get_feature_names_out(['Work_Mode', 'Location'])
encoded_df = pd.DataFrame(encoded_data, columns=new_columns, index=df.index)
df = pd.concat([df.drop(columns=['Work_Mode', 'Location']), encoded_df], axis=1)
print(df.head())
print(f"Number of new columns created: {len(new_columns)}")
```

Q7. Build Preprocessing Pipeline
Create a pipeline that:
Applies encoding to categorical columns
Applies scaling to numerical columns
Use:
ColumnTransformer
Pipeline

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
numerical_cols = ['Age', 'Hours_Worked_Per_Week', 'Projects_Completed', 'Performance_Score', 'Salary']
categorical_cols = ['Department', 'Gender', 'Work_Mode', 'Location']
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
pipeline
```

Q8. Apply Pipeline
Apply the pipeline on the dataset.
Display:
Transformed dataset
Shape of final dataset

```python
df = pd.read_csv('employee_productivity_dataset.csv')
processed_data = pipeline.fit_transform(df)
cat_feature_names = pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_cols)
all_feature_names = numerical_cols + list(cat_feature_names)
df_processed = pd.DataFrame(processed_data, columns=all_feature_names)
print("--- Transformed Dataset ---")
print(df_processed.head())
print("--- Shape of Final Dataset ---")
print(df_processed.shape)
```

Q9. Conceptual Question
Why is scaling important in Python?

Scaling is critical in machine learning because most algorithms rely on mathematical calculations (like calculating the distance between data points). If your dataset has features with vastly different ranges—such as a Performance_Score between 1 and 5, and a Salary between 40,000 and 100,000—the algorithm will unfairly prioritize the larger numbers. The model would treat Salary as much more important simply because of its magnitude. Scaling (like Standardization or Normalization) forces all numerical features onto a similar scale, ensuring every feature contributes equally to the model's predictions and allowing algorithms to process the data much faster.

Q10. Conceptual Question
Why do we convert categorical data into numerical form?

Machine learning models are fundamentally mathematical equations; they cannot read or understand text strings like "Remote", "Finance", or "Delhi". They only understand numbers. We must convert categorical data into a numerical format (using techniques like One-Hot Encoding or Label Encoding) to translate those text values into a language the algorithm can process, calculate, and ultimately use to find patterns.
